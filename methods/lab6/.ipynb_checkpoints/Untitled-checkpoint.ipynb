```python
num = 1e-3
y0 = np.zeros(11)
y1 = np.zeros(11)
y2 = np.zeros(11)
y3 = np.zeros(11)
for n in range(5, 16):
    A = np.array([[0.0] * n] * n)
    for i in range(0, n):
        for j in range(0, n):
            A[i][j] = math.tan(i + 1) ** (17 - j)

    d = np.array([28] * n)
    d = mult(A, d)
    A[0][0] = A[0][0] + num
    
    y0[n-5] = count_pogr(A, num)[0]
    y1[n-5] = count_fact_pogrLU(A, d, num)
    y2[n-5] = count_fact_pogrLUch(A, d, num)
    y3[n-5] = count_fact_pogrLib(A, d, num)
```

```python
print_vec_hor(y1, '{:.12f}')
print_vec_hor(y2, '{:.12f}')
print_vec_hor(y3, '{:.12f}')
```
